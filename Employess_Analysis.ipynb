**Note:**

Here we Remove Outlier which is present in Salary column which is 200000.


```python
Q1 = df['Salary'].quantile(0.25)
Q3 =df['Salary'].quantile(0.73)

IQR = Q3-Q1

lower_bound = Q1 -1.5*IQR
upper_bound = Q3 +1.5*IQR

df =df[(df['Salary']>=lower_bound) & (df['Salary'] <=upper_bound)]

print("\nCleaned data (outliers removed):\n", df)
```

**1. What is the distribution of employess salaries? (Histogram/KDE plot)**

```python
plt.figure(figsize = (8,5))

# Set the theme before plotting
sns.set_style('whitegrid')

sns.histplot(data=df, x='Salary', kde=True, color ='teal', palette='Set2')
plt.title('Distribution of Employee Salaries')
plt.xlabel('Salary')
plt.ylabel('Number of employees(Frequency)')
plt.show()
```

**2.How many employes are in each department ?**

```python
plt.figure(figsize = (6,4))
sns.countplot(x = 'Department', data = df, palette='Set2')
plt.title('Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees(Frequancy)')
plt.show()
```

**What is the average salary by department? (Bar Chart(group Average))**

```python
plt.figure(figsize =(7,4))
sns.barplot(x = 'Department', y = 'Salary', data = df, palette='Set2')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()
```

**what are the top most common remarks ?**

```python
remarks_count = df['Remarks'].value_counts()
remarks_count_df = remarks_count.reset_index()
remarks_count_df.columns = ['Remarks', 'Count']
```

```python
plt.figure(figsize=(7, 4))
sns.barplot(x='Count',y='Remarks',data=remarks_count_df,palette='viridis')
plt.title('Remarks Count Distribution')
plt.xlabel('Count')
plt.ylabel('Remarks')
plt.show()
```
